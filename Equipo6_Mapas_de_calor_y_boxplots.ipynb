## **Boxplots**

###GYC

Para una mejor visualización de los datos, eliminamos los outliers de estas gráficas con cuartiles y las reemplazamos con el promedio de la columna completa. El objetivo de esto es poder visualizar de manera apropiada donde se encuentran "una mayoría" de los datos por columna, pero también se generó el que cuenta con los datos completos para poder contar con una alternativa con todos los datos.

##### Sin outliers

```python
gyc_copy = gyc.copy()
gyc_copy = gyc_copy.drop(gyc_copy[gyc_copy['NIVEL_GASTO'] == 'DEVOLUCIÓN'].index)
gyc_sinoutliers = gyc_copy[['IMPORTE', 'TOTAL MX', 'TOTAL SAT', 'AÑO']]
gyc_sinoutliers.set_index('AÑO', inplace=True)

q=gyc_sinoutliers[['IMPORTE', 'TOTAL MX', 'TOTAL SAT']]
percentile25=q.quantile(0.25) #Q1
percentile75=q.quantile(0.75) #Q3
iqr= percentile75 - percentile25
Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
gyc_sinoutliers = gyc_sinoutliers[(q<Limite_Superior_iqr)&(q>Limite_Inferior_iqr)]

gyc_sinoutliers = gyc_sinoutliers.fillna(round(gyc_sinoutliers.mean(),2))
```

```python
importe = px.box(gyc_sinoutliers, x=gyc_sinoutliers.index, y='IMPORTE')
importe.update_layout(title="Monto pagado de importe, por transacción")
importe.update_layout(xaxis_title="Año", yaxis_title="Importe")
importe.update_layout(plot_bgcolor='#f8f8f8')
importe.update_traces(marker=dict(color='red'))
importe.show()
```

```python
totalsat = px.box(gyc_sinoutliers, x=gyc_sinoutliers.index, y='TOTAL SAT', color=gyc_sinoutliers.index)
totalsat.update_layout(title="Monto pagado de TOTAL SAT, por transacción")
totalsat.update_layout(xaxis_title="Año", yaxis_title="Total SAT")
totalsat.update_layout(plot_bgcolor='#f8f8f8')
totalsat.update_traces(marker=dict(color='red'))
totalsat.show()
```

```python
totalmx = px.box(gyc_sinoutliers, x=gyc_sinoutliers.index, y='TOTAL MX', color=gyc_sinoutliers.index)
totalmx.update_layout(title="Monto pagado de TOTAL MX, por transacción")
totalmx.update_layout(xaxis_title="Año", yaxis_title="Total en MX (por transacción)")
totalmx.update_layout(plot_bgcolor='#f8f8f8')
totalmx.update_traces(marker=dict(color='red'))
totalmx.show()
```

## **Preguntas y Conclusiones**

**¿Hay alguna variable que no aporta información?**

**Si tuvieras que eliminar variables, ¿cuáles quitarías y por qué?**

Claro que a nuestros para nosotros hay muchas variables que no nos aportan mucho para el análisis de los datos. En previas entregas nosotros eliminamos las que veíamos que no brindaban un propósito además de ayudar a mantener un control interno y solo mantuvimos las que nos permitían hacer operaciones útiles en el tablero. Las columnas que fueron eliminadas del DataFrame inicial que nos proporcionó el socio formador fueron:

* Columna etiquetada como 'RFC' de la base de datos de FACTURACIÓN.
* Columnas 'TIP_DOC N', 'RFC' y 'FOLIO N' de la base de datos de notas.
* Columnas 'TIP_DOC N', 'RFC', 'SERIE D' y 'FOLIO D de la base de datos de devoluciones.
* Columna etiquetada como 'DESCRIPCIÓN' de la base de datos detalles22.
* Columnas etiquetadas como 'Columna1' y 'GASTO' de la base de datos gyc20(gastos y costos 2020).
* Columna etiquetada como 'MP' de la base de datos gyc21(gastos y costos 2021).
* Columnas etiquetadas como 'MP' y 'OTROS' de la base de datos gyc22(gastos y costos 20202.
* Columnas etiquetadas como 'MP', 'OTROS', y 'FP' de la base de datos gyc23(gastos y costos 2023).


La manera en la que elegimos que columnas eliminar fue diferente dependiendo cada DataFrame, por ejemplo en los de facturación, notas y devoluciones, consideramos que para el calculo de las ventas no es muy relevante tener todas las fechas de cada documento, con tener la fecha de elaboración es suficiente, así mismo, si se cuenta con el número de clientes, no creemos que es vital tener el rfc también.

Por otra parte, en el Dataframe de detalles, para la información financiera que buscamos, consideramos que la descripción del producto esta de más.

Finalemente, para los DataFrames de gastos y costos, buscamos igualarlos, es decir que estén compuestos por las mismas columnas para concatenarlos más adelante, por lo tanto, se eliminaron aquellas columnas que estuvieran solo en uno de los Dataframes o que faltara en alguno.

**¿Existen variables que tengan datos extraños?**

Además de valores faltantes y claves que solo aparecían un par de ocasiones, los únicos valores que nos llamaron la atención al estar tratando con las bases de datos fueron los  valores negativos que encontrábamos en el documento de facturación. Más adelante nos percatamos que este era el caso de alguna venta a la cual se le había aplicado una devolución. Esto fue tratado durante la etapa inicial de selección y limpieza y cuando generamos variables más adelante para poder categorizar las transacciones por relevancia de la venta, se les asignó su categoría para poder incluirlas en nuestras operaciones.
